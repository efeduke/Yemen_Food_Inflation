
Yemen, a country located on the tip of the Arabian Peninsula has been caught up in a conflict since 2015. This ongoing war has resulted in a crisis impacting aspects of life for Yemeni citizens, including the economy. One of the problems arising from this crisis is the rise, in food prices.

Importance of the Issue:

The skyrocketing costs of food have become a concern affecting millions of people and leading to increased levels of malnutrition and hunger. In an environment where over 80% of the population relies on aid understanding the dynamics, behind food inflation is not an economic matter but also an urgent humanitarian priority.

Objective:

The goal of this project is to examine the patterns, causes and consequences of food inflation in Yemen. By employing data driven approaches this research aims to provide insights that can benefit policymakers, humanitarian organizations and other stakeholders involved in mitigating this crisis.

```r
Yemen <- food[food$country == "Yemen" , ]
```

```r
library(forecast)

#create a time series object
yem <- ts(data= Yemen$Inflation ,start= 2009,frequency =12)
yem
```

Here are my paraphrased observations:

1. Limited Data, in 2009; We can see that there is a gap in the data for the year 2009 as there is no line segment representing that year.
2. Fluctuating Inflation; The graph displays fluctuations in inflation rates especially starting from 2015. During this period we observe both low rates of inflation reflecting the unstable economic conditions within the country.
3. Spikes in Inflation; Notable increases in inflation can be observed in the years 2015, 2018 and 2021. These spikes may indicate events or crises that took place during those years.
4. Stabilization; Interestingly the data for 2023 shows a decline and stabilization of inflation rates with some instances showing negative values. This could be an indication of recovery or other factors, at play that warrant investigation.



```r
library(repr)
options(repr.plot.width=15, repr.plot.height = 5)

#overview of the data
autoplot(yem)
```

Key Observations:

Initial Data Gap; It is noticeable that there are no data points for the year 2009 as there is no line representing that year on the graph.

Volatility and Patterns; The graph clearly shows a significant amount of volatility particularly starting from 2015 onwards. This volatility reflects the conditions, in Yemen, which have been worsened by ongoing conflicts.

Highs and Lows; We can observe spikes in inflation during years such as 2015, 2018 and 2021. These spikes might coincide with events or crises that had an impact on food prices.

Recent Patterns; The latest available data for 2023 indicates a trend of decreasing or negative inflation rates. This could suggest some level of stabilization although further investigation would be necessary to understand the underlying reasons, for this trend.
